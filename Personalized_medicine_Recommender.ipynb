```python
import numpy as np
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = pd.read_csv('medicine.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.duplicated().sum()
```

```python
df['Description']
```

```python
df['Description'].apply(lambda x:x.split())
```

```python
df['Reason'] = df['Reason'].apply(lambda x:x.split())
df['Description'] = df['Description'].apply(lambda x:x.split())
```

```python
df['Description']=df['Description'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tags'] = df['Description']  + df['Reason']
```

```python
new_df = df[['index','Drug_Name','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df
```

Natural language processing
using nltk library

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
def stem (text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(medicine):
  medicine_index = new_df[new_df['Drug_Name'] == medicine].index[0]
  distances = similarity[medicine_index]
  medicine_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
  for i in medicine_list:
    print(new_df.iloc[i[0]].Drug_Name)
```

```python
recommend("Paracetamol 125mg Syrup 60mlParacetamol 500mg Tablet 10'S")
```
